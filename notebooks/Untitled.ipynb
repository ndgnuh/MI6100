```python
import cv2
from PIL import Image
import numpy as np
```

```python
sift = cv2.SIFT_create()
```

```python
img = cv2.imread("./../samples/box.png")
img = imagegray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
keypoints = sift.detect(img, None)
output_image = cv2.drawKeypoints(imagegray, keypoints, 0, (0, 255, 0), flags=cv2.DRAW_MATCHES_FLAGS_DEFAULT)
Image.fromarray(output_image)
```
